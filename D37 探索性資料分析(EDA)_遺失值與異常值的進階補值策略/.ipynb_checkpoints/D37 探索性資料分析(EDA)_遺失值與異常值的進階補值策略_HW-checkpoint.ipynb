## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
df_train = pd.read_csv("Titanic_train.csv")
df_train.head()
```

## 觀察 Age

```python
g = sns.catplot(x="Sex", y="Age", hue="Sex",
               data=df_train, kind="box")
```

男生年紀比女生大

## 觀察 Pclass

```python
g = sns.catplot(x="Pclass", y="Age", hue="Pclass",
               data=df_train, kind="box")
```

年紀越高的會住在Pclass1；年紀越小的會住在Pclass3

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_train['Sex']=le.fit_transform(df_train['Sex'])
df_train = df_train.join(pd.get_dummies(df_train['Embarked'], dummy_na=True, prefix='Embarked'))
df_train.drop(['Name', 'PassengerId', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
display(df_train)
```

```python
from sklearn.impute import KNNImputer
value_neighbors=3
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(df_train))
#display(data[missing_index.values==True])
#display(df_filled[missing_index.values==True])
df_filled.columns = df_train.columns
display(df_filled)
df_filled.isna().sum()
```
