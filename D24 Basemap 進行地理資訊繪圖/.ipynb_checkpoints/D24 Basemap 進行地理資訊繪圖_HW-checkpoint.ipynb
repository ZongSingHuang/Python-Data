 當 "projection" 參數為“Mill"時，所得平展位面

```python
# 載入套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖。
# 當 "projection" 參數為“ortho"時，所得圖位地球儀截面
'''
#由此開始寫code 
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
'''

#設定圖形大小
plt.figure(figsize=(8, 8))

#設定投影型態
m = Basemap(projection='ortho', resolution=None, lat_0=50, lon_0=-100)

#設定圖形顏色, 選用NASA 寶石藍輸出 
m.bluemarble(scale=0.5)

#秀圖
plt.show()
```

```python
'''
由此開始寫code 

#設置Lambert保形底圖。
# width=12000000,height=9000000
#經緯度lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
#set resolution = None跳過邊界數據集的處理
# projection='lcc'
'''
m = Basemap(width=12000000, height=9000000, projection='lcc',
            resolution=None, lat_1=45., lat_2=55, lat_0=50, lon_0=-107.)

#選用NASA 寶石藍輸出 
m.bluemarble()

#秀圖
plt.show()
```

```python
'''
由此開始寫code 
#  "projection" 參數為“Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90,
# 左下角的經度 - llcrnrlon = -180,
# 右上角的緯度 - urcrnrlat = 90,
# 右上角的經度 -urcrnrlon = 180

# 成圖設定resolution
'''
m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180)

m.drawcoastlines()
m.drawcounties(color='darkred')
m.drawstates(color='b')
m.drawcounties()

plt.title('Basemap Tutorial')
plt.show()
```
